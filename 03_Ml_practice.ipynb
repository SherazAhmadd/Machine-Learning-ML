# Simple Linear regression
- Input variable --> feature --> Independent variable --> Train variable
- output variable --> predictions --> Dependent variable--> test variable

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# importing file 
df = pd.read_csv('Salary_Data.csv')
df.head()
# Splitting the data 
x = df[['YearsExperience']]
y = df[['Salary']]
# fitting into Linear Regression
reg = LinearRegression().fit(x,y)
reg
```

```python
# y = a + bx ---> LinearRegression Equation
# a = constant/intercept
# b = function / slope of x
print("coffiecient in (a) Linear Regression Equation",reg.coef_)
print("coffiecient in (b) Linear Regression Equation",reg.intercept_)
```

```python
x.head()
```

```python
print("The predicted salary on 1-Year experience is: ",reg.predict([[1]]))
```

## Now train the model and then testing and evaluating the score 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2 ,random_state= 0)
print("model score",reg.score(x,y))
print("test score",reg.score(x_test,y_test))
print("train score",reg.score(x_train,y_train))
reg.predict([[1]])
```

### Plotting the test and train models

```python
plt.scatter(x_train,y_train)
plt.xlabel("Year experience")
plt.ylabel("Salary")
plt.title("plot of train")
# Creating the linear Line
plt.plot(x_train,reg.predict(x_train),color= "green")
plt.show()
```

```python
plt.scatter(x_test,y_test)
plt.xlabel("Year experience")
plt.ylabel("Salary")
plt.title("plot of test")
plt.plot(x_test,reg.predict(x_test),color="black")
plt.show()
```

# Multiple Linear Regression

```python
df2 = pd.read_csv('Salary_Data2.csv')
df2.head()
```

```python
# splitting the data
x1 = df2[['YearsExperience','Age']]
y1 = df2[['Salary']]
```

```python
# Training the model
x1_train,x1_test,y1_train,y1_test = train_test_split(x1,y1,test_size = 0.2, random_state = 0)
```

```python
# fitting into the LinearRegression
reg1= LinearRegression().fit(x1,y1)
reg1
```

```python
# Prediction the unknown Value's in model bases
print("predicted salary is: ",reg1.predict([[1.1,21.0]]))
```

# Plotting 

```python
# getting predicted value because the attributes are not same
y_pred = reg1.predict(x1_test)
plt.scatter(y1_test,y_pred)
```
