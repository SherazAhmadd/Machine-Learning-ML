## Focusing on images ananlysis in Logistic Regression 

```python
import pandas as pd
import numpy as np
```

```python
# import data from online
from  sklearn.datasets import load_digits
digits = load_digits()
```

```python
# means 1797 pictures and their size is 64 means 8x8 
x = digits.data
print("so called x",digits.data.shape)
y = digits.target
print("this is our y",digits.target.shape)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(20,4))
for index,(image,label) in enumerate(zip(digits.data[0:10],digits.target[0:10])):
    plt.subplot(1,10,index +1 )
    plt.imshow(np.reshape(image,(8,8)),cmap=plt.cm.gray)
    plt.title(label,fontsize=20)
```

```python
# splitting the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
# checking the data it splits or not
print("Train input data",x_train.shape)
print("Test inputdata",x_test.shape)
print("Train output data",y_train.shape)
print("Test output data",y_test.shape)
```

```python
# Train the data
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression().fit(x_train,y_train)
log_reg
```

```python
# now predicting the Value
log_reg.predict(x_test[0:10])
predicted_values = log_reg.predict(x_test)
predicted_values
```

```python
# checking the accuracy
score = log_reg.score(x_test,y_test)
print("The accuracy score is: ",int(score*100),"%")
```

```python
# Confusion matrics
from sklearn import metrics
cm = metrics.confusion_matrix(y_test,predicted_values)
cm
```

```python
# creating heatmap of above
import seaborn as sns
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square=True,cmap='Spectral')
plt.xlabel('Actual output')
plt.ylabel('predicted output')
all_sample_title= 'accuracy score: {0}'.format(score)
plt.title(all_sample_title,size=15)
```
